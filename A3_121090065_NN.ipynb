```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
```

```python
train = pd.read_csv("train.csv")
new_columns = ['pixel'+str(x)for x in range(1,785)]
X_train = train.iloc[:,2: ]
X_train.columns = new_columns
y_train = train["Category"]
X_train.head()
```

```python
test = pd.read_csv("test.csv")
X_test = test.iloc[:, 1:]
y_test = test["label"]
X_test.head()
```

```python
layers = [1,2,3]
nodes = [50,200,784]
opt = ["adam","sgd"]
rates = [0.001, 0.005, 0.01]
```

```python
for layer in layers:  
    for node in nodes:
        for optimizer in opt:
            for rate in rates: 
                model = MLPClassifier(hidden_layer_sizes=[node for x in range(layer)],solver=optimizer,learning_rate='constant',learning_rate_init=rate).fit(X_train,y_train)
                print("When # of layers =",layer,", # of hidden nodes =",node,", the optimizer is",optimizer,", the learning rate =",rate,"\n",
                      "the train accuracy is %.4f" % model.score(X_train,y_train),", the test accuracy is %.4f" %model.score(X_test,y_test),"\n")
    
                
```
